**SOLUTION**

**STEP-1**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Retail_sales_data.csv')
print(data.head(10))
```

**STEP-2**

```python
# replacing missing values with mean
numeric_cols = ["Quantity Sold", "Unit Price", "Total Sales Amount"]
for col in numeric_cols:
    data[col] = data[col].fillna(data[col].mean())

# replacing missing values with mode
categorical_cols = ["Payment Method", "Customer Segment"]
for col in categorical_cols:
    data[col] = data[col].fillna(data[col].mode()[0])

# Converting Date column to datetime
data["Date"] = pd.to_datetime(data["Date"])
print(data.dtypes)
```

```python
print("\nMissing values after replacement:")
print(data.isnull().sum())
```

**STEP-3**

```python
# Most common payment methods
payment_counts = data["Payment Method"].value_counts()
print('Most Common Payment Price 💳 :')
print(payment_counts)

# Cities with highest number of orders
city_order_counts = data["City"].value_counts()
print('Cities With Highest Number of Orders 🌆 :')
print(city_order_counts)
```

**STEP-4**

```python
# Group and summarize product sales
product_summary = (
    data.groupby("Product Name")
    .agg({"Unit Price": "mean", "Quantity Sold": "sum"})
    .reset_index()
)
product_summary.rename(columns={
    "Unit Price": "Avg_Unit_Price",
    "Quantity Sold": "Total_Quantity_Sold"
}, inplace=True)
product_summary["Avg_Unit_Price"] = product_summary["Avg_Unit_Price"].round(4)
product_summary["Total_Quantity_Sold"] = product_summary["Total_Quantity_Sold"].round(4)
print(product_summary.head())
```

```python
# Category purchased frequency by customer
category_counts = (
    data.groupby(["Customer Segment", "Category"])
    .size()
    .reset_index(name="Purchase_Count")
)
category_counts_sorted = category_counts.sort_values(
    by=["Customer Segment", "Purchase_Count"],
    ascending=[True, False]
)
print(category_counts_sorted)
```

**STEP-5**

```python
# Total spending by each customer
customer_spending = (
    data.groupby("Customer ID")["Total Sales Amount"].sum()
    .reset_index()
)
customer_spending["Total Sales Amount"] = customer_spending["Total Sales Amount"].round(4)
customer_spending = customer_spending.sort_values(
    by="Total Sales Amount", ascending=False
)
print(customer_spending.head(10))
```

```python
# Number of  orders per customer
customer_orders = (
    data.groupby("Customer ID")["Order ID"].nunique()
    .reset_index()
)
customer_orders.rename(columns={"Order ID": "Total_Orders"}, inplace=True)
repeat_buyers = (customer_orders["Total_Orders"] > 1).sum()
one_time_buyers = (customer_orders["Total_Orders"] == 1).sum()
print("Repeat/Regular Buyers:", repeat_buyers)
print("One-Time Buyers:", one_time_buyers)
print(customer_orders.head(10))
```

**STEP-6**

```python
daily_orders = data.groupby('Date')['Order ID'].nunique().reset_index()
plt.figure(figsize=(12,6))
sns.lineplot(x='Date', y='Order ID', data=daily_orders, marker='o',color='purple')
plt.title(" Daily Order Volume Over Time", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
product_revenue = data.groupby('Product Name')['Total Sales Amount'].sum().reset_index()
top_products = product_revenue.sort_values(by='Total Sales Amount', ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x='Total Sales Amount', y='Product Name', data=top_products, palette="viridis")
plt.title(" Top 10 Products by Total Revenue", fontsize=14)
plt.xlabel("Total Revenue")
plt.ylabel("Product Name")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x="Customer Segment", data=data, palette="viridis", legend=False)
plt.title("Orders by Customer Segment", fontsize=14)
plt.xlabel("Customer Segment")
plt.ylabel("Number of Orders")
plt.tight_layout()
plt.show()
```

**STEP-7**

```python
product_revenue = data.groupby('Product Name')['Total Sales Amount'].sum().reset_index()
top_products = product_revenue.sort_values(by='Total Sales Amount', ascending=False).head(10)
top_products.to_csv("Top_10_Products_Revenue.csv", index=False)
print(" Top 10 Products by Revenue exported as 'Top_10_Products_Revenue.csv'")
```

```python
country_sales = data.groupby('Country').agg({
    'Total Sales Amount': 'sum',
    'Order ID': 'nunique'
}).reset_index()
country_sales['Average_Order_Value'] = (country_sales['Total Sales Amount'] / country_sales['Order ID']).round(2)
country_sales.to_csv("Country_AOV.csv", index=False)
print(" Country-wise AOV exported as 'Country_AOV.csv'")
```
